## Dataset Overview

**Data Description :**  <br>
The "marketing_campaign" dataset is used to analyze customer behavior and responses to marketing campaigns. It includes various customer attributes and campaign-specific data.<br>


**Features :**  <br>
* **ID**: Unique identifier for each customer.<br>
* **Year_Birth**: The birth year of the customer.<br>
* **Education**: The education level of the customer (e.g., Basic, Secondary, Tertiary).<br>
* **Marital_Status**: The marital status of the customer (e.g., Single, Married).<br>
* **Income**: The income of the customer.<br>
* **Kidhome**: Number of children living at home.<br>
* **Teenhome**: Number of teenagers living at home.<br>
* **Dt_Customer**: Date when the customer was acquired.<br>
* **Recency**: Number of days since the customer’s last purchase.<br>
* **MntWines**: Amount spent on wine products.<br>
* **MntFruits**: Amount spent on fruit products.<br>
* **MntMeatProducts**: Amount spent on meat products.<br>
* **MntFishProducts**: Amount spent on fish products.<br>
* **MntSweetProducts**: Amount spent on sweet products.<br>
* **MntGoldProds**: Amount spent on gold products.<br>
* **NumDealsPurchases**: Number of purchases with discounts.<br>
* **NumWebPurchases**: Number of purchases made online.<br>
* **NumCatalogPurchases**: Number of purchases made through catalogs.<br>
* **NumStorePurchases**: Number of purchases made in-store.<br>
* **NumWebVisitsMonth**: Number of web visits in the last month.<br>
* **AcceptedCmp3**: Whether the customer accepted the 3rd campaign (1 = Yes, 0 = No).<br>
* **AcceptedCmp4**: Whether the customer accepted the 4th campaign (1 = Yes, 0 = No).<br>
* **AcceptedCmp5**: Whether the customer accepted the 5th campaign (1 = Yes, 0 = No).<br>
* **AcceptedCmp1**: Whether the customer accepted the 1st campaign (1 = Yes, 0 = No).<br>
* **AcceptedCmp2**: Whether the customer accepted the 2nd campaign (1 = Yes, 0 = No).<br>
* **Complain**: Whether the customer complained about the campaign (1 = Yes, 0 = No).<br>
* **Z_CostContact**: Cost of contact with the customer.<br>
* **Z_Revenue**: Revenue from the customer.<br>
* **Response**: Whether the customer responded positively to the campaign (1 = Yes, 0 = No).<br>

## Clustering

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()


df['Education'] = label_encoder.fit_transform(df['Education'])
df['Marital_Status'] = label_encoder.fit_transform(df['Marital_Status'])
df['Dt_Customer'] = label_encoder.fit_transform(df['Dt_Customer'])
df['Customer_Age'] = label_encoder.fit_transform(df['Customer_Age'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
col_scale = ['Income','Recency', 'Complain', 'Response', 'Family_Size',
       'Depenses', 'TotalAcceptedCmp', 'NumTotalPurchases']

df[col_scale] = scaler.fit_transform(df[col_scale])
```

```python
df.head(10)
```

```python
df_copy = df.copy()
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df_copy)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(16,8))
plt.plot(range(1,11),wcss, 'bx-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### K-means


```python
# Training a predicting using K-Means Algorithm.

kmeans=KMeans(n_clusters=2, random_state=42).fit(df_copy)
pred=kmeans.predict(df_copy)


# Appending those cluster value into main dataframe (without standard-scalar)

df_copy['cluster_Kmeans'] = pred + 1

df_copy.head()
```

```python
from sklearn.metrics import davies_bouldin_score

# data_array est votre ensemble de données
# predictions sont les clusters prédits par K-means

db_score = davies_bouldin_score(df_copy, pred)
print("Davies-Bouldin Score:", db_score)
```

**Score de silhouette :** mesure la distance moyenne entre chaque instance et le cluster le plus proche dont elle ne fait pas partie. Les valeurs vont de -1 à 1, où une valeur élevée indique que l’objet est bien adapté à son propre cluster et mal adapté aux clusters voisins.

Un score de ***Davies-Bouldin*** de **0.5177** est considéré comme assez bon, car il est proche de zéro. Cela suggère que les clusters dans votre modèle sont bien séparés et qu’il y a une dispersion faible à l’intérieur des clusters1. C’est un bon indicateur que cet modèle K-means a effectué un clustering efficace des données.

```python
from sklearn.metrics import silhouette_score

# data_array est votre ensemble de données
# predictions sont les clusters prédits par K-means

silhouette_avg = silhouette_score(df_copy, pred)
print("Silhouette Score:", silhouette_avg)
```

***Un score de silhouette*** de **0.6143** est considéré comme assez bon. Cela indique que les clusters sont raisonnablement bien définis et que les échantillons au sein de chaque cluster sont plus proches les uns des autres par rapport aux échantillons des autres clusters3. En d’autres termes, ce modèle K-means semble avoir fait un bon travail en séparant les différents clusters.

**Indice de Calinski-Harabasz :** également connu sous le nom de score de variance ratio, il est basé sur la dispersion entre les clusters et la dispersion dans les clusters.

```python
from sklearn.metrics import calinski_harabasz_score

calinski_harabasz = calinski_harabasz_score(df_copy, pred)
print("Calinski-Harabasz Score:", calinski_harabasz)
```

Un score de **6390.61** est considéré comme très élevé, ce qui suggère que les clusters formés par votre modèle K-means sont très bien définis et séparés. Les clusters ont une dispersion interne faible et une bonne séparation les uns des autres, ce qui indique une partition de haute qualité

### k-means from scratch

```python
import numpy as np
import pandas as pd

class KMeansFromScratch:
    def __init__(self, n_clusters, max_iter=300):
        self.n_clusters = n_clusters
        self.max_iter = max_iter

    def fit(self, X):
        # Step 1: Initialize centroids
        centroids = X[np.random.choice(X.shape[0], self.n_clusters, replace=False)]

        for _ in range(self.max_iter):
            # Step 2: Assign points to clusters
            clusters = [[] for _ in range(self.n_clusters)]
            for point in X:
                distances = [np.linalg.norm(point - centroid) for centroid in centroids]
                cluster_idx = np.argmin(distances)
                clusters[cluster_idx].append(point)

            # Step 3: Update centroids
            new_centroids = [np.mean(cluster, axis=0) for cluster in clusters]

            # Step 4: Check convergence
            if np.allclose(centroids, new_centroids):
                break

            centroids = new_centroids

        self.centroids = centroids
        self.clusters = clusters

    def predict(self, X):
        distances = np.array([[np.linalg.norm(point - centroid) for centroid in self.centroids] for point in X])
        return np.argmin(distances, axis=1)

# Créer un sous-ensemble de données contenant les fonctionnalités souhaitées
selected_features = ['Education', 'Marital_Status', 'Income',
       'Dt_Customer', 'Recency', 'Complain', 'Response', 'Family_Size',
       'Depenses', 'TotalAcceptedCmp', 'NumTotalPurchases', 'Customer_Age']
subset_df = df[selected_features]

# Convertir les colonnes numériques du sous-ensemble de données en un tableau NumPy
numeric_data_array = subset_df.select_dtypes(include=[np.number]).values

# Exemple d'utilisation
kmeans = KMeansFromScratch(n_clusters=2)
kmeans.fit(numeric_data_array)
predictions = kmeans.predict(numeric_data_array)
print(predictions)
```

```python
from sklearn.metrics import davies_bouldin_score

# data_array est votre ensemble de données
# predictions sont les clusters prédits par K-means

db_score = davies_bouldin_score(df, predictions)
print("Davies-Bouldin Score:", db_score)
```
